# Circuit Theorems

To handle the complexity of electric power systems, some theorems to simplify circuit
analysis has been developed.
These theorems are applicable to linear circuits, we first discuss the
concept of circuit linearity.

Important aspects following circuit linearity:
 - superposition
 - source transformation
 - maximum powertransfer
 -  Thevenin’s theorem
 -  Norton’s theorem

## Circuit Linearity

When there is a linear relationship between cause and effect we have a linear property.

```
A linear circuit is one whose output is linearly related (or directly proportional)
to its input.
```

For **ohms law**:

$$ v = i R $$

we can determine linerity if the constant *k* applied to the current causes a corresponding voltage difference:

$$ kv = ki R $$

The power equation:

$$ p = i^2R =  \frac{v^2}{R} $$

is a quadratic function, hence non-linear, and the linear property does not apply.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 10  # Ohms

# Current range
I = np.linspace(0, 10, 200)

# Ohm's Law
V = R * I

# Power
P_i = I**2 * R         # P = I²R
P_v = V**2 / R         # P = V²/R

# Create side-by-side plots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Linear relationship (Ohm's Law)
ax[0].plot(I, V, color='blue', linewidth=2)
ax[0].set_title("Ohm's Law: V = RI")
ax[0].set_xlabel("Current I [A]")
ax[0].set_ylabel("Voltage V [V]")
ax[0].grid(True)

# Plot 2: Quadratic relationship (Power)
ax[1].plot(I, P_i, color='red', linewidth=2, label=r'$P=I^2R$')
ax[1].plot(I, P_v, '--', color='black', linewidth=1.5, label=r'$P=V^2/R$')
ax[1].set_title("Power Relationship")
ax[1].set_xlabel("Current I [A]")
ax[1].set_ylabel("Power P [W]")
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
````
