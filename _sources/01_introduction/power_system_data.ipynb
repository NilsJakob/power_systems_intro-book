# Power System Data

To get knowledge of the power system we can acess open data sources from the Transmission System Operators (TSO). In  Norway the TSO is Statnett. The TSO operates the transmission system and is responsible for transmitting and balancing power and maintaining and developing the transmission infrastructure. Statnett gathers [data from the power system](https://www.statnett.no/en/for-stakeholders-in-the-power-industry/data-from-the-power-system/) and provide an overview of how the power system operates. Statnett is part of the European Network of Transmission Sytem Operators ([ENTSO-E](https://www.entsoe.eu/)), together with 40 other TSO's from Europe. On ENTSO-E [transparency platform](https://transparency.entsoe.eu/) european power system data is published.
