```python
# FEEDBACK CELL

import numpy as np

def check(student, expected, name):
    if expected is None:
        print(f"{name}: cannot evaluate yet")
        return False
    elif student is None:
        print(f"{name}: no answer yet")
        return False
    elif isinstance(student, (int, float)):
        if abs(student - expected) < 1e-6:
            print(f"{name}: correct")
            return True
        else:
            print(f"{name}: incorrect")
            return False
    elif isinstance(student, np.ndarray):
        if np.allclose(student, expected):
            print(f"{name}: correct")
            return True
        else:
            print(f"{name}: incorrect")
            return False
    elif isinstance(student, tuple):
        if student == expected:
            print(f"{name}: correct")
            return True
        else:
            print(f"{name}: incorrect")
            return False
    else:
        print(f"{name}: invalid output")
        return False

results = []
t = np.linspace(0, 0.1, 100)

# Q1
def sol_signal(t, V_hat, f, theta):
    return V_hat * np.cos(2*np.pi*f*t + theta)
results.append(check(voltage_signal(t,325,50,0), sol_signal(t,325,50,0), "Q1"))

# Q4
results.append(check(per_unit(10,5), 2.0, "Q4"))

# Q5
results.append(check(compute_rms(325), 325/np.sqrt(2), "Q5"))

# Q7
results.append(check((voltage_unit,current_unit,power_unit), ('Volt','Ampere','Watt'), "Q7"))

print('--- Summary ---')
print(f"Score: {sum(results)}/{len(results)}")
```
