# Buils vs. Buy

- The **orange** line (*Buying an off-the-shelf solution*): Illustrates the ongoing effort when choosing a proprietary, ready-to-use product. Initially, effort and costs are low as the system is ready to deploy. Over time, however, time and resource consumption increase drastically because the company is tied to the vendor's framework (vendor lock-in), making it highly demanding and expensive to add new, customized functionality.
- The **blue** line (*Building from scratch*): Illustrates the trajectory if you invest in developing a proprietary solution from the ground up. This requires a large investment and high effort in the initial phase. Over time, however, the effort decreases significantly because the system is built on open standards and is interoperable. This makes it far easier and more cost-effective to further develop functionality later (payback).

```python
# TEST CELL
import numpy as np
import matplotlib.pyplot as plt

# Time axis
x = np.linspace(0, 10, 500)

# Interoperable software curve
y_interoperable = (
    0.55
    + 0.15 * np.exp(-((x - 3) / 2.5) ** 2)
    - 0.45 / (1 + np.exp(-(x - 6)))
)

# Vendor locked software curve
y_vendor = (
    0.18
    + 0.65 / (1 + np.exp(-(x - 7)))
)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(
    x,
    y_interoperable,
    #color="#2b7a78",
    color='blue',
    linewidth=2,
    label="Interoperable software"
)

ax.plot(
    x,
    y_vendor,
    color="orange",
    linewidth=2,
    label="Vendor locked software"
)

ax.annotate(
    "Development\nfrom scratch\n(Investment)",
    xy=(3, 0.7),
    xytext=(3, 0.9),
    arrowprops=dict(arrowstyle="->", color="#8aa5a5"),
    color="#4f6f6f",
    ha="center",
    fontsize=12
)

ax.annotate(
    "Buy common\nvendor software",
    xy=(2.6, 0.17),
    xytext=(1.7, 0.03),
    arrowprops=dict(arrowstyle="->", color="blue"),
    fontsize=12
)

ax.annotate(
    "Add\nfunctionality",
    xy=(9.2, 0.82),
    xytext=(9.2, 1.0),
    arrowprops=dict(arrowstyle="->", color="gray"),
    ha="center",
    fontsize=12
)

ax.annotate(
    "Add\nfunctionality\n(Payback)",
    xy=(9.2, 0.15),
    xytext=(9.2, 0.25),
    arrowprops=dict(arrowstyle="->", color="#8aa5a5"),
    #color="#2b7a78",
    ha="center",
    fontsize=12
)

ax.text(
    10.3,
    y_vendor[-1],
    "Vendor locked\nsoftware",
    fontsize=13,
    va="center"
)

ax.text(
    10.15,
    y_interoperable[-1] - 0.02,
    "Interoperable\nsoftware",
    fontsize=13,
    color="#2b7a78",
    #color='blue',
    va="center"
)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_xlabel("Time", fontsize=16, fontweight="bold")
ax.set_ylabel(
    "Development and usage effort",
    fontsize=16,
    fontweight="bold",
    rotation=90,
    labelpad=5
)

ax.set_xlim(0, 12.5)
ax.set_ylim(0, 1.05)

ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()
plt.show()
```
