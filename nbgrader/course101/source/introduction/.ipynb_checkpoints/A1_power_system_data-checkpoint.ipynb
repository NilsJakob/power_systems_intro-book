# Assignment A0: Power System Data (Norway)

This assignment is based on real data from the Norwegian power system.

## Learning objectives

- Interpret system-level energy data (TWh)
- Perform basic calculations
- Understand energy mix and surplus
- Reflect on system implications



## Given Data (2025)

| Variable              | Value                  |
|----------------------|------------------------|
| Production           | 162 TWh               |
| Consumption          | 139 TWh               |
| Hydropower share     | 90%                   |
| Wind power share     | 9%                    |
| Solar share          | 0.2%                  |
| Thermal share        | 1%                    |

```python
score = 0
total = 0
```

```python
# Data provided
production = 162
consumption = 139

hydro_share = 0.90
wind_share = 0.09
solar_share = 0.002
thermal_share = 0.01
```

## Q1: Power balance

Define a variable `surplus` and compute the power surplus:

Surplus = Production − Consumption

Return the result in TWh.

```python
### BEGIN SOLUTION
surplus = production - consumption

# hidden grading test
assert abs(surplus - (production - consumption)) < 1e-6
### END SOLUTION
```


## Q2: Renewable share

Hydropower and wind power are considered renewable.

Compute the total renewable share (as a fraction)

Store your result in the variable `renewable_share`.

```python
### BEGIN SOLUTION
renewable_share = hydro_share + wind_share

# hidden grading test
expected = hydro_share + wind_share
assert abs(renewable_share - expected) < 1e-6
### END SOLUTION
```


## Q3: Hydropower production

Compute the total hydropower production in TWh, in a variable `hydro`.


```python
### BEGIN SOLUTION
hydro = production * hydro_share

# hidden grading test
expected = production * hydro_share
assert abs(hydro - expected) < 1e-6
### END SOLUTION
```

## Q4: Non-renewable share

Estimate the non-renewable share of the system, in a variable named `non_renewable`.



```python
### BEGIN SOLUTION
renewable_share = hydro_share + wind_share + solar_share
non_renewable = 1 - renewable_share

# hidden grading test
expected = 1 - (hydro_share + wind_share + solar_share)
assert abs(non_renewable - expected) < 1e-6
### END SOLUTION
```


## Q5: Wind power production

Compute wind power production in TWh.

Use variable `wind`


```python
### BEGIN SOLUTION
wind = production * wind_share

# hidden grading test
expected = production * wind_share
assert abs(expected - expected) < 1e-6
### END SOLUTION
```

```python
def check(student, expected, name):
    if student is None:
        print(f"{name}: no answer yet")
        return False
    elif abs(student - expected) < 1e-6:
        print(f"{name}: correct")
        return True
    else:
        print(f"{name}: incorrect")
        return False
```

```python
results = []

results.append(check(surplus, production - consumption, "Q1 surplus"))
results.append(check(renewable_share, hydro_share + wind_share + solar_share, "Q2 renewable_share"))
results.append(check(hydro, production * hydro_share, "Q3 hydro"))
results.append(check(non_renewable, 1 - renewable_share, "Q4 non_renewable"))
results.append(check(wind, production * wind_share, "Q5 wind"))

score = sum(results)
total = len(results)
```

```python
print("\n Final score:")
print(f"{score}/{total} correct")

if score == total:
    print("Perfect!")
elif score >= total * 0.7:
    print("Good job!")
else:
    print("Keep practicing and revise your answers")
```
