# Assignment A0: Power System Data (Norway)

This assignment is based on real data from the Norwegian power system.

## Learning objectives

- Interpret system-level energy data (TWh)
- Perform basic calculations
- Understand energy mix and surplus
- Reflect on system implications



## Given Data (2025)

| Variable              | Value                  |
|----------------------|------------------------|
| Production           | 162 TWh               |
| Consumption          | 139 TWh               |
| Hydropower share     | 90%                   |
| Wind power share     | 9%                    |
| Solar share          | 0.2%                  |
| Thermal share        | 1%                    |

```python
score = 0
total = 0

def check_answer(student, expected, name, hint=""):
    global score, total
    total += 1
    
    if student is None:
        print(f"{name}: no answer yet")
    elif abs(student - expected) < 1e-6:
        print(f"{name} is correct")
        score += 1
    else:
        print(f"{name} is incorrect")
        if hint:
            print(f"Hint: {hint}")
```

```python

# Data provided
production = 162
consumption = 139

hydro_share = 0.90
wind_share = 0.09
solar_share = 0.002
thermal_share = 0.01
```

## Q1: Power balance

Compute the power surplus:

Surplus = Production − Consumption

Return the result in TWh.

```python

# YOUR CODE HERE
surplus = None
```

```python
# SOLUTION
surplus = production - consumption
surplus

```

```python
### BEGIN HIDDEN TEST
assert abs(surplus - (production - consumption)) < 1e-6
### END HIDDEN TEST

```


## Q2: Renewable share

Hydropower and wind power are considered renewable.

Compute the total renewable share (as a fraction)


```python

# YOUR CODE HERE
renewable_share = None
```

```python
# SOLUTION
renewable_share = hydro_share + wind_share
renewable_share
```

```python
# TEST
expected = hydro_share + wind_share
assert abs(renewable_share - expected) < 1e-6

```


## Q3: Hydropower production

Compute the total hydropower production in TWh.


```python

# YOUR CODE HERE
hydro = None
```

```python
# SOLUTION
hydro = production * hydro_share
hydro
```

```python
# TEST
expected = production * hydro_share
assert abs(hydro - expected) < 1e-6
```

## Q4: Non-renewable share

Estimate the non-renewable share of the system.

Hint: Sum all renewable sources first.


```python

# YOUR CODE HERE
renewable = None
```

```python
# SOLUTION
renewable = hydro_share + wind_share + solar_share
non_renewable = 1 - renewable
non_renewable
```

```python
# TEST
expected = 1 - (hydro_share + wind_share + solar_share)
assert abs(non_renewable - expected) < 1e-6
```


## Q5: Wind power production

Compute wind power production in TWh.


```python

# YOUR CODE HERE
wind = None
```

```python
# SOLUTION
wind = production * wind_share
wind
```

```python
# TEST
expected = production * wind_share
assert abs(expected - expected) < 1e-6
```

```python

print("Checking all answers...\n")

check_answer(surplus, production - consumption, "Q1 surplus")
check_answer(renewable_share, hydro_share + wind_share, "Q2 renewable share")
check_answer(hydro, production * hydro_share, "Q3 hydro")
check_answer(non_renewable, 1-renewable, "Q4 non-renewable share")
check_answer(wind, production * wind_share, "Q5 wind power")
```

```python
print("🔍 Checking all answers...\n")

check_answer(surplus, production - consumption, 
             "Q1 surplus", 
             "Subtract consumption from production")

check_answer(renewable_share, hydro_share + wind_share, 
             "Q2 renewable share", 
             "Add hydro and wind shares")

check_answer(hydro, production * hydro_share, 
             "Q3 hydropower", 
             "Multiply production by hydro share")

check_answer(non_renewable, 1 - renewable, 
             "Q4 non-renewable share", 
             "Subtract renewable share from 1")

check_answer(wind, production * wind_share, 
             "Q5 wind power", 
             "Multiply production by wind share")
```
