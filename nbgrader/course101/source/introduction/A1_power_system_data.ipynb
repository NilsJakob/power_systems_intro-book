## Given Data (2025)

| Variable              | Value                  |
|----------------------|------------------------|
| Production           | 162 TWh               |
| Consumption          | 139 TWh               |
| Hydropower share     | 90%                   |
| Wind power share     | 9%                    |
| Solar share          | 0.2%                  |
| Thermal share        | 1%                    |

```python

# Data provided
production = 162
consumption = 139

hydro_share = 0.90
wind_share = 0.09
solar_share = 0.002
thermal_share = 0.01
```

## Q1: Power balance

Compute the power surplus:

Surplus = Production − Consumption

Return the result in TWh.

```python

# YOUR CODE HERE
raise NotImplementedError()
```

```python
# SOLUTION
surplus = production - consumption
surplus

```

```python
# TEST
expected = production - consumption
assert abs(surplus - expected) < 1e-6

```


## Q2: Renewable share

Hydropower and wind power are considered renewable.

Compute the total renewable share (as a fraction)


```python

# YOUR CODE HERE
raise NotImplementedError()
```

```python
# SOLUTION
renewable_share = hydro_share + wind_share
renewable_share
```

```python
# TEST
expected = hydro_share + wind_share
assert abs(renewable_share - expected) < 1e-6

```


## Q3: Hydropower production

Compute the total hydropower production in TWh.


```python

# YOUR CODE HERE
raise NotImplementedError()
```

```python
# SOLUTION
hydro = production * hydro_share
hydro
```

```python
# TEST
expected = production * hydro_share
assert abs(hydro - expected) < 1e-6
```

## Q4: Non-renewable share

Estimate the non-renewable share of the system.

Hint: Sum all renewable sources first.


```python

# YOUR CODE HERE
raise NotImplementedError()
```

```python
# SOLUTION
renewable = hydro_share + wind_share + solar_share
non_renewable = 1 - renewable
non_renewable
```

```python
# TEST
expected = 1 - (hydro_share + wind_share + solar_share)
assert abs(non_renewable - expected) < 1e-6
```
