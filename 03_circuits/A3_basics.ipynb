```python
# FEEDBACK CELL
def check(student, expected, name):
    if student is None:
        print(f"{name}: no answer yet")
        return False
    elif abs(student - expected) < 1e-3:
        print(f"{name}: correct")
        return True
    else:
        print(f"{name}: incorrect")
        return False

results = []

# correct values
omega = 2 * np.pi * 50
XL_true = omega * 0.05
XC_true = 1 / (omega * 100e-6)
X_true = XL_true - XC_true
Z_true = np.sqrt(10**2 + X_true**2)
phi_true = np.degrees(np.arctan(X_true / 10))

results.append(check(XL, XL_true, "Q1 XL"))
results.append(check(XC, XC_true, "Q2 XC"))
results.append(check(X, X_true, "Q3 X"))
results.append(check(Z, Z_true, "Q4 Z"))
results.append(check(phi, phi_true, "Q5 phi"))

score = sum(results)
total = len(results)
```
